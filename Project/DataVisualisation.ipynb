```python
kolumny = ['CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5']
liczba = df[kolumny].stack().value_counts()

plt.bar(liczba.index, liczba.values)
plt.xlabel('Wartość')
plt.ylabel('Ilość wystąpień')
plt.title('Ilość wystąpień unikalnych wartości w pięciu kolumnach')
plt.xticks(rotation=90)
plt.show()
```
