```python
# LOGIKA DANYCH
ranni = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF MOTORIST INJURED']
zabici = ['NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST KILLED']
czynniki = ['CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5']

df['INJURED'] = df[ranni].sum(axis=1)
df['KILLED'] = df[zabici].sum(axis=1)

def polacz_wartosci(row):
    wartosci = row.dropna().astype(str)
    wartosci = [w for w in wartosci if w != 'Unspecified']
    return ', '.join(wartosci)

polaczone_wartosci = df[czynniki].apply(polacz_wartosci, axis=1)

liczba_unikalnych = polaczone_wartosci.nunique()

# MNHATAN
df_manhattan = df[df['BOROUGH'] == 'MANHATTAN'].copy()  # Używamy metody copy() dla df_manhattan

df_manhattan.loc[:, 'INJURED'] = df_manhattan[ranni].sum(axis=1)  # Używamy .loc[row_indexer, col_indexer]
df_manhattan.loc[:, 'KILLED'] = df_manhattan[zabici].sum(axis=1)  # Używamy .loc[row_indexer, col_indexer]

licznik_ranni = df_manhattan[ranni].apply(polacz_wartosci, axis=1).value_counts()
licznik_zabici = df_manhattan[zabici].apply(polacz_wartosci, axis=1).value_counts()

# 5 najwyższych wartości dla "czynniki"
top_ranni = licznik_ranni.nlargest(5)
top_zabici = licznik_zabici.nlargest(5)

# Generowanie wykresu
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(12, 20))

# ZŁY WYKRES!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!


# blok wykresu
for i, (value_ranni, count_ranni) in enumerate(top_ranni.items()):
    value_zabici = top_zabici.index[i]
    count_zabici = top_zabici.iloc[i]
    
    ax = axes[i]
    ax.bar(['Ranni', 'Zabici'], [count_ranni, count_zabici], color=['blue', 'red'])
    ax.set_title(f'Czynnik: {value_ranni} ({value_zabici})')

fig.tight_layout()
plt.show()
```

```python
# TWORZENIE LOGIKI ZBIORÓW

ranni = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF MOTORIST INJURED']
zabici = ['NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST KILLED']
czynniki = ['CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5']

# sumy:
df['INJURED'] = df[ranni].sum(axis=1)
df['KILLED'] = df[zabici].sum(axis=1)



def polacz_wartosci(row):
    wartosci = row.dropna().astype(str)
    wartosci = [w for w in wartosci if w != 'Unspecified']
    return ', '.join(wartosci)

# => połączone czynniki
polaczone_wartosci = df[czynniki].apply(polacz_wartosci, axis=1)

liczba_unikalnych = polaczone_wartosci.nunique()

# MANHATAN
df_manhattan = df[df['BOROUGH'] == 'MANHATTAN'].copy()

df_manhattan.loc[:, 'INJURED'] = df_manhattan[ranni].sum(axis=1)
df_manhattan.loc[:, 'KILLED'] = df_manhattan[zabici].sum(axis=1)

licznik = df_manhattan.groupby(polaczone_wartosci)[['INJURED', 'KILLED']].sum()
licznik = licznik[licznik.index != 'Unspecified']

# "NAJNIEBEZPIECZNIEJSZE" - ZAKŁĄDAMY 5 NAJCZEŚCIEJ POWODUJĄCYCH OBRAŻENIA ZESTAWÓW CZYNNIKÓW
top_5 = licznik.sum(axis=1).nlargest(5)

# /////////////////////////////////////////////
# BLOK WYKRES
fig, ax = plt.subplots(figsize=(18, 10))

width = 0.35
bar1 = ax.bar(range(5), licznik.loc[top_5.index, 'INJURED'], width, label='Ranni')
bar2 = ax.bar([x + width for x in range(5)], licznik.loc[top_5.index, 'KILLED'], width, label='Zabici')

ax.set_xticks([x + width / 2 for x in range(5)])
ax.set_xticklabels(top_5.index, rotation=90)
ax.set_xlabel('Czynniki')
ax.set_ylabel('Liczba przypadków')
ax.set_title('Liczba rannych i zabitych dla najczęstszych czynników w dzielnicy Manhattan')
ax.legend()

plt.tight_layout()
plt.show()

# /////////////////////////////////////////////
# BLOK TESTOWY

print("Największe sumy dla połączonych czynników:")
for czynnik, suma in top_5.items():
    print(f"{czynnik}: {suma}")
    
```

```python
# ACCIDENT DATE na typ daty i DAY OF WEEK z nazwami dni tygodnia
df['ACCIDENT DATE'] = pd.to_datetime(df['ACCIDENT DATE'])
df['DAY OF WEEK'] = df['ACCIDENT DATE'].dt.day_name()

daily_counts = df['DAY OF WEEK'].value_counts()

daily_counts = daily_counts.sort_index()

# WYKRES
plt.figure(figsize=(10, 6))
sns.barplot(x=daily_counts.index, y=daily_counts.values, palette='viridis')
plt.xlabel('Dzień tygodnia')
plt.ylabel('Liczba wypadków')
plt.title('Ilość wypadków z podziałem na dni tygodnia')
plt.show()
```

```python
# ACCIDENT DATE na typ daty i DAY OF WEEK z nazwami dni tygodnia
df['ACCIDENT DATE'] = pd.to_datetime(df['ACCIDENT DATE'])

# MIESIĄC I ROK
selected_month = 8  # Wybrany miesiąc (np. sierpień)
selected_year = 2019  # Wybrany rok

filtered_df = df[(df['ACCIDENT DATE'].dt.month == selected_month) & (df['ACCIDENT DATE'].dt.year == selected_year)]
filtered_df['DAY OF WEEK'] = filtered_df['ACCIDENT DATE'].dt.day_name()

# powtarzające się ID 'COLLISION_ID'
df_unique_ids = filtered_df.drop_duplicates(subset=['COLLISION_ID'])
daily_counts = df_unique_ids['DAY OF WEEK'].value_counts().sort_index()

# WYKRES
plt.figure(figsize=(10, 6))
sns.barplot(x=daily_counts.index, y=daily_counts.values, palette='viridis')
plt.xlabel('Dzień tygodnia', fontweight='bold', color='darkblue')
plt.ylabel('Liczba wypadków', fontweight='bold', color='darkblue')
plt.title(f'Ilość wypadków z podziałem na dni tygodnia - {selected_month}/{selected_year}')

plt.show()
```
